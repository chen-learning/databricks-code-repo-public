## Daa Conversion from Structured to Semi-Structure
##### How to convert csv to json?
- read teh csv file then write to json
  - spark.read.csv().write.json()


```python
# How to convert csv to json
## csv read
print("read as csv")
spark.read.csv(path="/Volumes/workspace/wd36schema/ingestion_volume/source/custs_header",header=True,inferSchema=True).show(2)

## convert to json
spark.read.csv(path="/Volumes/workspace/wd36schema/ingestion_volume/source/custs_header",header=True,inferSchema=True).write.json("/Volumes/workspace/wd36schema/ingestion_volume/target/jsonout_custs", mode='overwrite')

## json read
print("converted json")
spark.read.json("/Volumes/workspace/wd36schema/ingestion_volume/target/jsonout_custs").show(2)
```
