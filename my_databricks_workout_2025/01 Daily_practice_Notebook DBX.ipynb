```python
dbutils.widgets.text("tablename","cities","Enter the table name:")
```

```python
tbl_name = dbutils.widgets.get("tablename")
print(tbl_name)
```

```python
df = spark.sql(f"select * from default.{tbl_name}")
display(df)
```

```python
dbutils.widgets.dropdown("name_list", "Chenthil",["Chenthil", "Arun", "Balaji"],"Select your name:")
```
