##1.Wrting inside Databricks environment itself
Ingestion (read data from cloud storage into databricks cluster) <br>
Curation (Transform) <br>   
**Egress (write into internal / external storage with/without schema migration/data conversion)** 

####schema migration / structure change - may involve below operations
- header change 
>selected columns |
>newly derived columns |
>change in column name |
>change in column order |
>change in data type

- sep/delimeter change
- mode

#### Structured data (2D) - CSV
Tables / data frames with rows and columns

```python
ingest_df1 = spark.read.csv("/Volumes/workspace/wd36schema/ingestion_volume/source/custs_header", header=True, sep=',', inferSchema=True,samplingRatio=0.1)
ingest_df1.printSchema()
display(ingest_df1.limit(5)) 
```

```python
ingest_df1.write.csv(path="/Volumes/workspace/wd36schema/ingestion_volume/target/csvout")
# write without header
```

```python
ingest_df1.write.csv(path="/Volumes/workspace/wd36schema/ingestion_volume/target/csvout",sep='~', header=True)
```

```python
# mode (4) - append, overwrite, ignore, error
ingest_df1.write.csv(path="/Volumes/workspace/wd36schema/ingestion_volume/target/csvout",sep='~', header=True, mode='overwrite')
```

```python
transformed_df = ingest_df1.select("custid", "fname", "lname", "profession", "age").withColumnRenamed("profession", "prof")
# DSL (Domain Specific Language) way of transformation
transformed_df.write.csv("/Volumes/workspace/wd36schema/ingestion_volume/target/csvout",sep='~', header=True, mode='overwrite')
```

```python
transformed_df.write.csv("/Volumes/workspace/wd36schema/ingestion_volume/target/csvout_compressed",sep='~', header=True, mode='overwrite',compression='gzip')
```

#### Semi Structured / dynamic schema -JSON

```python
transformed_df.write.json("/Volumes/workspace/wd36schema/ingestion_volume/target/jsonout", mode='overwrite')
```

```python
ingest_df1.select("custid","age","profession").write.json("/Volumes/workspace/wd36schema/ingestion_volume/target/jsonout", mode='append')
```
